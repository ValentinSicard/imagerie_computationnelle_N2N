# Creation des filtres Ã  la main et comparaison avec le modÃ¨le d'apprentissage sur notre Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
def mse(image1, image2):
    err = np.sum((image1 - image2) ** 2) / image1.size
    return err

def psnr(image1, image2):
    mse_val = mse(image1, image2)
    if mse_val == 0:
        return float('inf')
    return 20 * np.log10(1 / np.sqrt(mse_val))
```

```python
path = "C:/ai/images_size/"
path_noise = "C:/ai/images_noise/"
path_filter = "C:/ai/images_filter/"

dirs = os.listdir( path )
print(dirs)

psnr_noise_list = []
psnr_bilfilter_list = []

for item in dirs:
    if os.path.isfile(os.path.expanduser(path+item)):
        print(path+item)

        im_clean = plt.imread(os.path.expanduser(path+item))
        im = plt.imread(os.path.expanduser(path_noise+item))
        psnr_noise = psnr(im_clean, im)
        psnr_noise_list.append(psnr_noise)

        im_float = np.float32(im)
        im_bilfilter = cv.bilateralFilter(im, 9, 75, 75)
        psnr_bilfilter = psnr(im_clean, im_bilfilter)
        psnr_bilfilter_list.append(psnr_bilfilter)

        plt.imsave(path_filter + item + ' _bilfilt' + 'av' + "{:.2f}".format(psnr_noise) + 'ap' + "{:.2f}".format(psnr_bilfilter) + '.png', im_bilfilter)

print("psnr noise mean :", np.mean(psnr_noise_list))
print("psnr bilfilter mean :", np.mean(psnr_bilfilter_list))
```

```python
path = "C:/ai/images_size/"
path_noise = "C:/ai/images_noise/"
path_filter = "C:/ai/images_filter/"

dirs = os.listdir( path )
print(dirs)

psnr_noise_list = []
psnr_filter_list = []

for item in dirs:
    if os.path.isfile(os.path.expanduser(path+item)):
        print(path+item)

        im_clean = plt.imread(os.path.expanduser(path+item))
        im = plt.imread(os.path.expanduser(path_noise+item))
        psnr_noise = psnr(im_clean, im)
        psnr_noise_list.append(psnr_noise)


        im_filter = bm3d.bm3d(im, sigma_psd=30/255)
        im_filter = np.minimum(np.maximum(im_filter, 0), 1)
        print(im_filter)
        psnr_filter = psnr(im_clean, im_filter)
        psnr_filter_list.append(psnr_filter)

        plt.imsave(path_filter + item + ' _bm3d' + 'av' + "{:.2f}".format(psnr_noise) + 'ap' + "{:.2f}".format(psnr_filter) + '.png', im_filter)

print("psnr noise mean :", np.mean(psnr_noise_list))
print("psnr bm3d mean :", np.mean(psnr_filter_list))
```
